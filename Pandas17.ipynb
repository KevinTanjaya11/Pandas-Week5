```python
import pandas as pd
import numpy as np

print(pd.__version__)
print(np.__version__)
#import modul
```

```python
n_rows = 365 * 24
n_cols = 2
cols = ['col1', 'col2']

df = pd.DataFrame(np.random.randint(1, 20, size=(n_rows, n_cols)),
                 columns=cols)

df.index = pd.util.testing.makeDateIndex(n_rows, freq='H')
df

#persiapan data frame, kolom baris diisi 365 * 24, 24 mempresentasikan jumlah jam dalam sehari dan 365 mempresentasikan jumlah hari dalam satu tahun
#setiap colom diberi label 1 dan 2
#colom 1 dan colom 2 berisi nilai acak
```

```python
df.resample('M')['col1'].sum().to_frame()
#interval waktu berdasarkan bulan dengan method resample dengan memasukan paramater M yang meresampling bulan
```

```python
df.resample('D')['col1'].sum().to_frame()
#interval waktu berdasarkan perhari dengan method resample denagn memasukan paramater D yang mepresentasikan Daily dan menggunakan fungsi agregsi sum 
```
